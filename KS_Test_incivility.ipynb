## KS-Test for destributions of performance values

### Get Macro-Avg. f1-scores of all Train/test combinations (n= 12) and all runs (n= 20)

```python
#Store all values from data frames in a list, row 20 (aggregated value) is excluded
dflist = [list(df_low_low["f1"][:20]),
          list(df_low_med["f1"][:20]),
          list(df_low_high["f1"][:20]),
          list(df_low_mix["f1"][:20]),
          
          list(df_med_low["f1"][:20]),
          list(df_med_med["f1"][:20]),
          list(df_med_high["f1"][:20]),
          list(df_med_mix["f1"][:20]),
         
          list(df_high_low["f1"][:20]),
          list(df_high_med["f1"][:20]),
          list(df_high_high["f1"][:20]),
          list(df_high_mix["f1"][:20]),
         
         ]
```

### Get all combinations of two for the 2 sample KS test

```python
#Assign labels to the data frames for permutation overview

dflist_label = ["df_low_low", "df_low_med", "df_low_high", "df_low_mix",
                
                "df_med_low", "df_med_med", "df_med_high", "df_med_mix",
                
                "df_high_low", "df_high_med", "df_high_high", "df_high_mix",
         
         ]
```

```python
#Permutation: Get all combination of test combinations for 2-sample KS-test from the list of data frames
res = [(a, b) for idx, a in enumerate(dflist) for b in dflist[idx + 1:]]
```

```python
#Permutation: Get all combination of test combinations for 2-sample KS-test from the labels
res_label = [(a, b) for idx, a in enumerate(dflist_label) for b in dflist_label[idx + 1:]]
for i in res_label:
    print(str(i))
```

### Apply 2 sample KS test

```python
res_label[0] #First combination
```

```python
res_label[0][0] # Label of first data frame in first combination
```

```python
ks_2samp(res[0][0],res[0][1]) #KS test for first combination
```

```python
result = ks_2samp(res[0][0],res[0][1]) #KS test for first combination
print("pvalue= ", (result[1]))
print("pvalue rounded= ", round(result[1],2))
```

#### Loop to apply 2 sample KS test for all combinations

```python
#Prints only non-significant deviations between 2 destributions

for label,i in zip(res_label, res):

   
    for e in i:
        list1 = i[0]
        list2 = i[1]
        ks = ks_2samp(list1,list2)



    if (ks[1]) <= 0.5:
        #print(label)
        #print("ks_value", ks[0])
        #print("p_value", round(ks[1],2))
        pass
    else:
        print("No difference: ", label)
        print("p_value", round(ks[1],2))

```

```python
#Saves values of significant results in list
def ks_all(res, res_label):
    
    liste_labels = []
    liste_p = []
    liste_ks = []


    
    for label,i in zip(res_label, res):

   
        for e in i:
            list1 = i[0]
            list2 = i[1]
            ks = ks_2samp(list1,list2)


        if (ks[1]) <= 0.5:

            liste_labels.append(label)
            liste_ks.append(ks[0])
            liste_p.append(round(ks[1],2))
        else:
            pass


        
    return liste_labels, liste_p, liste_ks 
```

```python
a,b,c = ks_all(res, res_label)
```

```python
df_ks_results = pd.DataFrame(
    {'combination': a,
     'p_value': b,
     'ks': c
    })
```
